```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# 1. Load the dataset.
```

```python
df=pd.read_csv('students_cleaned.csv')
```

```python
df
```

```python
# 2. Do basic calculations:
```

```python
# a. Average score in Math, Science, English
avg_scores = {
    'Math': df['math_score'].mean(),
    'Science': df['science_score'].mean(),
    'English': df['english_score'].mean()
}
print("Average Scores:", avg_scores)
```

```python
# b. Count of students by Gender
gender_counts = df['gender'].value_counts()
print("Gender Counts:\n", gender_counts)
```

```python
# 3. Create 2 simple charts:
# a. Bar chart → Average scores per subject
# b. Pie chart → Number of students by Gender.
# c. Add title and labels.
```

```python


#  bar chart
plt.figure(figsize=(6,4))
sns.barplot(x=list(avg_scores.keys()), y=list(avg_scores.values()), palette="viridis")

#  labels and title
plt.title("Average Scores by Subject")
plt.xlabel("Subject")
plt.ylabel("Average Score")

plt.show()
```

```python
# pie chart
plt.pie(gender_counts, labels=gender_counts.index, autopct="%1.1f%%", startangle=140, colors=['#66b3ff','#ff9999'])
plt.title("Distribution of Students by Gender")
plt.show()
```
